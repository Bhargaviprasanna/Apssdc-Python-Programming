```python
import pandas as pd
import os
df=pd.DataFrame({'Roll Number':[10,20,30,40],'Marks':[99,100,45,23]})
df
```

```python
# read a csv file using pandas
dfp='DataFiles/income.csv'
def readDf(dfp):
    return pd.read_csv(dfp)
readDf(dfp)
```

```python
# print column values in a dataframe
df=readDf(dfp)
def printData(df):
    columns = df.columns
    for i in columns:
        print(i,end='    ')
    return
printData(df)
def printrowData(df,row_name):
    print()
    values=df.values
    for row in values:
#         if row_name in row:
        print()
        for item in row:
            print(item,end='  ')
    return 
printrowData(df,'Alabama')
```

```python
def getrowkeyindex(df,rowkey):
    values=df.values
    for i in range(len(values)):
        if values[i][0]==rowkey or values[i][1]==rowkey:
            return i
    return
getrowkeyindex(df,'Alabama')
def getallvaluesinarow(df,col_name):
    x = df[col_name].unique()
    for k in x:
        rindex=getrowkeyindex(df,k)
        print(k,'index = ',rindex)
    return x
getallvaluesinarow(df,'State')
```

```python
def getcolkeyindex(df,colkey):
    for i in range(len(df.columns)):
        if df.columns[i] == colkey:
            return i
    return 
getcolkeyindex(df,'State')
def getindexesofcolumns(df):
    columns=df.columns
    print(columns)
    for i in range(len(columns)):
        colindex=getcolkeyindex(df,columns[i])
        print(columns[i],'index =',colindex)
    return
getindexesofcolumns(df)
```

```python
# get average of all from 2005 to 2013
col_name='State'
fromind=getcolkeyindex(df,'2005')
toind=getcolkeyindex(df,'2013')
def average(df,row_name,fromind,toind):
    values=df.values
    s=0
    c=0
    for i in values:
        if row_name in i:
            for j in range(fromind,toind+1):
                s+=i[j]
                c+=1
#     print(s,'is sum',c,'is count')
    return s/c
def getavgofallstates(df,col_name):
    state_avg={}
    colvalues=df[col_name].unique()
    for c in colvalues:
        x=average(df,c,fromind,toind)
        state_avg[c]=x
    return state_avg
# average(df,'Alaska',fromind,toind)
getavgofallstates(df,col_name)
```

```python
# State with highest average income in the last three years(2011 -2013)
col_name='State'
fromind=getcolkeyindex(df,'2011')
toind=getcolkeyindex(df,'2013')
def average2(df,row_name,fromind,toind):
    values=df.values
    s=0
    c=0
    for i in values:
        if row_name in i:
            for j in range(fromind,toind+1):
                s+=i[j]
                c+=1
#     print(s,'is sum',c,'is count')
    return s/c
def getmax(state_avg):
    sa=state_avg.values()
    x=max(sa)
    for k in state_avg.keys():
        if x == state_avg[k]:
            return k
def getavgofallstates2(df,col_name):
    state_avg={}
    colvalues=df[col_name].unique()
    for c in colvalues:
        x=average2(df,c,fromind,toind)
        state_avg[c]=x
    y=getmax(state_avg)
    return y
getavgofallstates2(df,'State')
```
